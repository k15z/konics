# Konics: Drive API
Once again, we'll start by making a simple track. This time, however, we won't manually specify the location and what to look at. Instead, we'll work through the drive API which makes the steering more "car-like".

```python
from konics import *
import matplotlib.pyplot as plt

track = Track(size=256)
for i in range(10):
    track.add(Cone(-10, 20*i))
    track.add(Cone( 10, 20*i))
track.add(Cone(-15, 40, 0))
track.add(Cone( 15, 40, math.pi))
```

We can create a Drive instance from the track - note that multiple Drive instances are allowed to share the same track so beware of mutation - and render it. By default, we start a position (0, 0) facing north.

```python
drive = Drive(track)
plt.imshow(drive.render())
plt.show()
```

Now let's try driving forward. Right now, the default is to move by 5 units every time forward is called, but this will probably change in the future.

```python
drive.forward()
plt.imshow(drive.render())
plt.show()
```

We can also rotate/turn by some number of radians. The current implementation allows you to turn on-the-spot but, once again, this may change in the future.

```python
drive.rotate(0.2)
plt.imshow(drive.render())
plt.show()
```
