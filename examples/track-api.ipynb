# Konics: Track API
Let's start by designing a simple track. This track will have a double line of cones stretching out to infinity and two sideways cones indicating that your car should drive down the middle.

```python
from konics import *
import matplotlib.pyplot as plt

track = Track(size=512)
for i in range(10):
    track.add(Cone(-10, 20*i))
    track.add(Cone( 10, 20*i))
track.add(Cone(-15, 40, 0))
track.add(Cone( 15, 40, 180))
```

Now let's render it. We'll start at location (0,-10) which is behind the starting line and look towards (0, 100) which is far away in the distance.

```python
look_at = (0, 100)
location = (0, -10)
image = track.render(location, look_at)
plt.imshow(image)
plt.show()
```

We can move forwards by changing our location to (0, 10)... see how much closer the cones are now?

```python
look_at = (0, 100)
location = (0, 10)
image = track.render(location, look_at)
plt.imshow(image)
plt.show()
```

We can also turn by changing what we look at... let's take a closer look at one of the sideways cones.

```python
look_at = (30, 100)
location = (0, 10)
image = track.render(location, look_at)
plt.imshow(image)
plt.show()
```
